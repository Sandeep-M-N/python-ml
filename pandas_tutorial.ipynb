```python
from io import StringIO,BytesIO
```

```python
data=('col1,col2,col3\n'
      'x,1,22.0\n'
      'y,2,23.1\n'
      'z,3,24.3\n')
```

```python
pd.read_csv(StringIO(data))
```

```python
## read from specific columns

df=pd.read_csv(StringIO(data),usecols=['col2','col3'])
```

```python
df
```

```python
## specifying column data types
data=('a,b,c,d\n'
      '1,2,3,4\n'
      '5,6,7,8\n'
      '9,10,11,12\n')
```

```python
sf=pd.read_csv(StringIO(data),dtype=float)
```

```python
sf
```

```python
## we can also specify each columns a datatype
data=('a,b,c,d\n'
      '1,2,3,4\n'
      '5,6,7,8\n'
      '9,10,11,12\n')
```

```python
df=pd.read_csv(StringIO(data),dtype={'a':int,'b':float,'c':'Int64','d':object})
```

```python
df
```

```python
df['c']
```

```python
## check the data types
df.dtypes
```
